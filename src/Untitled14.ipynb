```python
import gradio as gr
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestClassifier
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.impute import SimpleImputer

def train_penguin_model():
    """
    Loads the UCI Heart Disease dataset and trains the Penguin Healthcare model.
    """
    print("System: Loading Penguin Healthcare database...")

    # Load Data
    try:
        df = pd.read_csv('heart_disease_uci.csv')
    except FileNotFoundError:
        return None, "Error: 'heart_disease_uci.csv' not found. Please upload the file."

    df['target'] = df['num'].apply(lambda x: 1 if x > 0 else 0)


    feature_cols = ['age', 'sex', 'cp', 'trestbps', 'chol']
    X = df[feature_cols]
    y = df['target']


    numeric_features = ['age', 'trestbps', 'chol']
    numeric_transformer = Pipeline(steps=[
        ('imputer', SimpleImputer(strategy='mean')),
        ('scaler', StandardScaler())
    ])

    categorical_features = ['sex', 'cp']
    categorical_transformer = OneHotEncoder(handle_unknown='ignore')

    preprocessor = ColumnTransformer(
        transformers=[
            ('num', numeric_transformer, numeric_features),
            ('cat', categorical_transformer, categorical_features)
        ])

    model_pipeline = Pipeline(steps=[
        ('preprocessor', preprocessor),
        ('classifier', RandomForestClassifier(n_estimators=100, random_state=42))
    ])

    model_pipeline.fit(X, y)
    print("System: Penguin AI Model ready.")

    return model_pipeline, None


ai_model, error_msg = train_penguin_model()


def diagnostic_request(age, sex, cp, trestbps, chol, service_status):

    if age is None:
        return "âš ï¸ Error: Please enter a valid Age."


    if not service_status:
        return "âš ï¸ Penguin Healthcare System Offline. Request queued."

    if ai_model is None:
        return "Error: Model not loaded."

e
    input_data = pd.DataFrame({
        'age': [age],
        'sex': [sex],
        'cp': [cp],
        'trestbps': [trestbps],
        'chol': [chol]
    })


    prediction = ai_model.predict(input_data)[0]
    prediction_prob = ai_model.predict_proba(input_data)[0][prediction]

    diagnosis = "POSITIVE (Risk Detected)" if prediction == 1 else "NEGATIVE (No Risk Detected)"

    return (
        f"ðŸ§ **Penguin Healthcare Diagnosis Report**\n"
        f"----------------------------------------\n"
        f"Result: **{diagnosis}**\n"
        f"Confidence: {prediction_prob:.2%}\n"
    )



# Inputs
inputs = [
    gr.Number(label="Patient Age", value=55),  # CHANGED: Slider -> Number Box
    gr.Radio(["Male", "Female"], value="Male", label="Sex"),
    gr.Dropdown(
        ["typical angina", "atypical angina", "non-anginal", "asymptomatic"],
        value="asymptomatic",
        label="Chest Pain Type"
    ),
    gr.Slider(90, 200, value=130, label="Resting BP (mm Hg)"),
    gr.Slider(100, 600, value=250, label="Cholesterol (mg/dl)"),
    gr.Checkbox(value=True, label="System Online?")
]

# Outputs
outputs = gr.Markdown(label="Report")

# Interface with updated Title
app = gr.Interface(
    fn=diagnostic_request,
    inputs=inputs,
    outputs=outputs,
    title="Penguin Healthcare",  # CHANGED: Main Page Title
    description="Authorized Clinical Access Only. Enter patient vitals below.",
    theme="soft",
    css="footer {visibility: hidden}" # Clean up the look slightly
)


app.launch(
    auth=("clinician", "hospital123"),
    auth_message="<h2 style='text-align:center'>Penguin Healthcare Login</h2>",
    debug=True
)
```
